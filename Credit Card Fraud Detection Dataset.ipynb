```python
#Data Preprocessing
#Scaling Features
from sklearn.preprocessing import StandardScaler
```

```python
df["Time"]= StandardScaler().fit_transform(df["Time"].values.reshape(-1,1))
df["Amount"]= StandardScaler().fit_transform(df["Amount"].values.reshape(-1,1))
```

```python
#Explatory Data Analysis
#UniVariate Analysis
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
sns.histplot(df["Time"], bins=50)
plt.title("Distribution of transaction Times")
plt.show()
```

```python
#Vibariate Analysis
sns.boxplot(x="Class", y="Amount", data=df)
plt.title("Transaction amount by class")
plt.show()
```

```python
#Correlation between Class and Amount
df[["Class", "Amount"]].corr()
```

```python
corr_matrix = df.corr()
plt.figure(figsize=(6, 6))
sns.heatmap(corr_matrix, cmap='coolwarm', vmin=-1, vmax=1, annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
#Time series analysisdf
df['Hour'] = df['Time'].apply(lambda x: x // 3600)
sns.countplot(x='Hour', hue='Class', data=df)
plt.title('Transaction Frequency by Hour')
plt.show()
```

```python
sns.countplot(x='Class', data=df)
plt.title('Class Distribution')
plt.show()
```
